#Q.3 scrape images of fruits,cakes,guitar,machine learning,cars.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common.exceptions import ElementNotVisibleException,ElementNotSelectableException
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
import os
import urllib

# Set up Selenium WebDriver
driver = webdriver.Chrome('chromedriver.exe')
# Define the keywords
keywords = ['fruits', 'cars', 'machine learning', 'guitar', 'cakes']
# Iterate over the keywords
for keyword in keywords:
    # Navigate to images.google.com
    driver.get('https://images.google.com/')
    
    #full screen
    driver.fullscreen_window()
    
    # Find the search bar and enter the keyword
    search_bar = driver.find_element(By.NAME, 'q')
    search_bar.send_keys(keyword)
    
    #two seconds delay
    time.sleep(2)
    
    # Find the search button and click it
    search_button = driver.find_element(By.CSS_SELECTOR, 'button[type="submit"]')
    search_button.click()
    
    # Scroll down to load more images
    for _ in range(3):
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(2)
    
    # Get the image elements
    image_elements = driver.find_elements(By.CSS_SELECTOR, 'img.rg_i')
    
    # Create a directory to save the images
    os.makedirs(keyword, exist_ok=True)
    
    # Scrape the images
    for i in range(10):
        try:
            # Again find the image elements each time
            image_elements = driver.find_elements(By.CSS_SELECTOR, 'img.rg_i')
            
            # Get the image URL
            image_url = image_elements[i].get_attribute('src')
            
            # Download the image
            image_path = f'{keyword}/image_{i+1}.jpg'
            urllib.request.urlretrieve(image_url, image_path)
            
            print(f'Saved: {image_path}')
        except Exception as e:
            print(f'Error scraping image {i+1}: {str(e)}')
            
    print()
    print(f'{keyword}: Scraped 10 images')

# Close the browser
driver.quit()
```

```python
brand_name = []
name_product = []
price = []
ret_exc = []
exp_del = []
avail = []
#pro_url = []

#scrapping the required details
start=0
end=3
for page in range(start,end):#for loop for scrapping 4 page

    try:
        bn = driver.find_elements(By.XPATH,"//*[@class='a-size-base-plus a-color-base']")
        for i in bn:
            brand_name.append(i.text)
    except NoSuchElementException:
        brand_name.append('-')

# Extracting price and saving it into price list
    try:
        p = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
        for i in p:
            price.append(i.text)
    except:
        price.append('-')
# Extracting product name and saving it into name_product list
    try:
        pn = driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
        for i in pn:
            name_product.append(i.text)
    except:
        name_product.append('-')
        


        
    nxt_button=driver.find_elements(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    #scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
```

```python
df_products=pd.DataFrame({'Brand_name': brand_name,
                'Price': price,
               'Product_name': name_product })
df_products
```
