```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://www.motor1.com/"

#initialize driver
driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)

time.sleep(5)

#send keys to search button
search_option = driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[3]/div/div/div/form/input")
search_option.send_keys("50 most expensive cars")

time.sleep(5)

#click on search button
search_btn = driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[3]/div/div/div/form/button[1]")
search_btn.click()

time.sleep(2)

#scroll down
driver.execute_script("window.scrollTo(0, 1000)")

time.sleep(5)

#click on option
option_click = driver.find_element(By.XPATH,"/html/body/div[3]/div[9]/div/div[1]/div/div/div[2]/div/div[1]")
option_click.click()

time.sleep(5)

#making empty list
car_name =[]
car_price=[]
car_price_1 =[]

#fetching name of the car
name = driver.find_elements(By.XPATH,"//h3[@class='subheader']")
for i in name:
    car_name.append(i.text)
    
#fetching price of the car
for k in range(4,102,2):
    price = driver.find_element(By.XPATH,'//*[@id="article_box"]/div[1]/div[2]/div[1]/p['+str(k)+']/strong')
    car_price.append(price)
    
for j in car_price:
    car_price_1.append(j.text)

# #making dataframe
df = pd.DataFrame({"Name":car_name[:len(car_price_1)], "Price":car_price_1})
df

```

```python
driver.get("https://www.jagranjosh.com/")

time.sleep(5)
#click on Gk button
button_gk = driver.find_element(By.XPATH,"/html/body/div/div[1]/div/div[1]/div/div[5]/div/div[1]/header/div[3]/ul/li[3]/a")
button_gk.click()

time.sleep(3)
driver.execute_script("window.scrollTo(0, 1500)") 

time.sleep(5)
#click Accept button pop-up
button_accept = driver.find_element(By.XPATH,"/html/body/div/div/div/div[3]/div/div[2]/div/div[3]/div/a[3]")
button_accept.click()

time.sleep(5)
#click on PM list
button_pm_list = driver.find_element(By.XPATH,"/html/body/div/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a")
button_pm_list.click()

time.sleep(5)
#fetching whole table
table = driver.find_element(By.XPATH, "//div[@class='table-box'][1]")

#fetching rows of table
rows = table.find_elements(By.TAG_NAME, "tr")

Name=[]
Born_dead=[]
Term_of_office=[]
Remarks=[]
for row in rows[1:]:
    cells = row.find_elements(By.TAG_NAME, "td")
    name = cells[0].text
    born_dead = cells[1].text
    term_of_office = cells[2].text
    remarks = cells[3].text
    
    Name.append(name)
    Born_dead.append(born_dead)
    Term_of_office.append(term_of_office)
    Remarks.append(remarks)
    
# Close the browser
driver.quit()
    
df = pd.DataFrame({"Name":Name, "Born_dead":Born_dead,"Term_of_office":Term_of_office,"Remarks":Remarks})
df
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.azquotes.com')
```

```python
time.sleep(5)

top_quotes_btn = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top_quotes_btn.click()
```

```python
Quotes = []
Authors = []
Quote_types = []

while len(Quotes) < 1000:
    # Scrape the quotes, authors, and types on the current page
    quotes = driver.find_elements(By.XPATH,"//a[@class='title']")
    for i in quotes:
        Quotes.append(i.text)
        
    authors = driver.find_elements(By.XPATH,"//div[@class='author']/a")
    for j in authors:
        Authors.append(j.text)
        
    quote_types = driver.find_elements(By.XPATH,"//div[@class='tags']")
    for k in quote_types:
        Quote_types.append(k.text)
        


df1= pd.DataFrame({"Quotes": Quotes, "Authors": Authors, "Type_of_Quotes":Quote_types})
df1
```
